```python
def get_state(l,centerline):
    
    x = centerline[0] + l * np.cos(centerline[2])
    y = centerline[1] + l * np.sin(centerline[2])
    theta = centerline[2]
    k = 1 / (1/centerline[3] + l)
    
    return [x,y,theta,k]
```
